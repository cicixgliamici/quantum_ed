# 03 — Noise & fidelity sweep

We will:
- create a target state |+>
- convert to density matrix
- apply depolarizing / dephasing noise
- measure fidelity vs noise strength


```python
import numpy as np
import matplotlib.pyplot as plt

from quantum_ed.states import normalize
from quantum_ed.density import rho_from_ket, fidelity_pure
from quantum_ed.channels import depolarize_rho, dephase_rho
```

## Target state |+>

```python
psi_plus = normalize(np.array([[1.0],[1.0]], dtype=complex))
rho = rho_from_ket(psi_plus)
```

## Sweep p ∈ [0,1]

```python
ps = np.linspace(0, 1, 51)
F_dep = []
F_deph = []

for p in ps:
    rho_dep = depolarize_rho(rho, p)
    rho_deph = dephase_rho(rho, p)
    F_dep.append(fidelity_pure(psi_plus, rho_dep))
    F_deph.append(fidelity_pure(psi_plus, rho_deph))

plt.plot(ps, F_dep, label='depolarizing')
plt.plot(ps, F_deph, label='dephasing')
plt.xlabel('p')
plt.ylabel('Fidelity to |+>')
plt.title('Noise vs fidelity')
plt.legend()
plt.grid(True)
plt.show()
```
