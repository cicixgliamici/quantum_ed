# 04 — CHSH violation (Bell state)

We compute CHSH correlations on |Φ+>.

We use observables with outcomes ±1.
A standard choice to reach 2√2 is:
- A0 = Z, A1 = X
- B0 = (Z + X)/√2
- B1 = (Z - X)/√2


```python
import numpy as np

from quantum_ed.states import ket0
from quantum_ed.gates import H, CNOT, kron_n, apply, X, Z
from quantum_ed.linalg import dagger
```

## Build |Φ+>

```python
psi = kron_n(ket0(), ket0())
psi = apply(kron_n(H, np.eye(2, dtype=complex)), psi)
psi = apply(CNOT, psi)
```

## Expectation E(A ⊗ B) = <ψ| (A⊗B) |ψ>

```python
def expect(op, ket):
    return float(np.real((dagger(ket) @ op @ ket).item()))

I = np.eye(2, dtype=complex)

A0 = Z
A1 = X
B0 = (Z + X) / np.sqrt(2)
B1 = (Z - X) / np.sqrt(2)

E00 = expect(kron_n(A0, B0), psi)
E01 = expect(kron_n(A0, B1), psi)
E10 = expect(kron_n(A1, B0), psi)
E11 = expect(kron_n(A1, B1), psi)

S = E00 + E01 + E10 - E11

print('E00, E01, E10, E11 =', E00, E01, E10, E11)
print('CHSH S =', S)
print('2*sqrt(2) =', 2*np.sqrt(2))
```

✅ You should see a violation: |S| > 2.
