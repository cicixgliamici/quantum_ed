# 01 — Qubit states & Bloch sphere

We will:
1. build qubit states
2. compute measurement probabilities
3. map a state to the Bloch sphere


```python
import numpy as np
import matplotlib.pyplot as plt

from quantum_ed.states import ket0, ket1, normalize, bloch_vector, state_from_bloch
from quantum_ed.measurement import probs_comp_basis
from quantum_ed.viz.bloch import plot_bloch_vector
```

## Basis states

The computational basis is \(|0\rangle, |1\rangle\).

```python
psi0 = ket0()
psi1 = ket1()

print('psi0 =\n', psi0)
print('psi1 =\n', psi1)
print('P(measure 0|psi0), P(1|psi0) =', probs_comp_basis(psi0))
print('P(measure 0|psi1), P(1|psi1) =', probs_comp_basis(psi1))
```

## Superposition

\(|+\rangle = (|0\rangle + |1\rangle)/\sqrt{2}\)

```python
psi_plus = normalize(np.array([[1.0],[1.0]], dtype=complex))
print('P(0), P(1) for |+> =', probs_comp_basis(psi_plus))
print('Bloch vector for |+>:', bloch_vector(psi_plus))
```

```python
ax = plot_bloch_vector(bloch_vector(psi_plus))
plt.show()
```

## From Bloch angles

Create a state from angles (theta, phi).

```python
theta = np.pi/3
phi = np.pi/4
psi = state_from_bloch(theta, phi)
print('psi =\n', psi)
print('Bloch vector:', bloch_vector(psi))
ax = plot_bloch_vector(bloch_vector(psi))
plt.show()
```
