# 02 — Bell states & entanglement

We will:
- build a Bell state from gates
- compute reduced density matrices (partial trace)
- observe that each subsystem is maximally mixed


```python
import numpy as np

from quantum_ed.states import ket0
from quantum_ed.gates import H, CNOT, kron_n, apply
from quantum_ed.density import rho_from_ket, partial_trace_two_qubits
```

## Build |Φ+> = (|00> + |11>)/√2

Circuit: apply H on first qubit, then CNOT.

```python
psi = kron_n(ket0(), ket0())
psi = apply(kron_n(H, np.eye(2, dtype=complex)), psi)
psi = apply(CNOT, psi)

amps = psi.flatten()
labels = ['00','01','10','11']
for lab, amp in zip(labels, amps):
    print(lab, amp)
```

## Density matrix and partial trace

For an entangled pure state, each single-qubit reduced state can be mixed.

```python
rho = rho_from_ket(psi)

rhoA = partial_trace_two_qubits(rho, keep=0)  # keep first qubit
rhoB = partial_trace_two_qubits(rho, keep=1)  # keep second qubit

print('rhoA=\n', rhoA)
print('rhoB=\n', rhoB)
print('Expected I/2 =\n', np.eye(2, dtype=complex)/2)
```

✅ You should see something very close to I/2.
